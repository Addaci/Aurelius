### **Europeana APIs**

**Description:** Europeana aggregates digitized cultural heritage content from European museums, galleries, libraries, and archives, offering access to millions of items such as books, artworks, and audiovisual materials. The API enables searching and retrieving metadata and thumbnails of these cultural heritage objects.

[Europeana Data Model:](https://europeana.atlassian.net/wiki/spaces/EF/pages/987791389/EDM+-+Mapping+guidelines) The Europeana Data Model (EDM in short) is a theoretical data model that allows data to be presented in different ways according to the practices of the various domains which contribute data to Europeana. To create a practical implementation Europeana has not used all the classes and properties defined in the EDM model. In particular, the ore:Proxy class is not included because it is created within Europeana using data provided according to the Guidelines. Additionally, for its internal working Europeana utilises a different set of classes and properties. There are several case studies that address particular aspects of using EDM.

**Search API:** is the easiest API to use and understand. It interacts with Europeana's data in much the same way as the [Europeana website](https://www.europeana.eu/en) does. You can search for keywords, and the API will return all records that match that keyword. You can refine your search with more advanced filters and advanced query syntax. You can choose to only return objects with certain copyright statements, or you can choose to return the results in a language of your choice. This means that with the Search API, you can get a response to the query: 'Give me all objects by Vermeer that are openly licensed and have high-resolution images.'

**Record API:** The Record API provides direct access to Europeana’s data and metadata, which is modelled using EDM. EDM is an open flexible data model used to capture the data and metadata about Cultural Heritage Objects (CHOs). The Record API is used to retrieve all of the data and metadata that relates to a single Cultural Heritage object, which will have a single Europeana ID. A Europeana ID is made up of a dataset number, and a record ID. For the object at this URL: https://www.europeana.eu/item/90402/RP_P_1984_87, the dataset ID is 90402, and the record ID is RP_P_1984_87. Both are findable by looking at the URL.

**IIIF API:** The Europeana APIs offer for IIIF covers the Presentation API (versions 2.1 & 3), the Content Search API (version 1) and make use of the IIIF Image API when made available by content providers. Additionally, Europeana’s IIIF offer also includes support for fulltext (such as transcriptions, translations of transcriptions, captions and subtitles) via the IIIF Fulltext API using an EDM extension for fulltext. These APIs follow the specifications defined by the IIIF consortium.

URL: https://www.europeana.eu/en/apis

# **3. Example: The UK National Archives's Discovery API**

**What is it?** The UK National Archives (TNA) API provides programmatic access to metadata and digital resources from the UK National Archives. It allows users to explore historical records, archival descriptions, and government documents preserved in the UK.

**What can it be used for?**

`Search and Retrieval:` Query metadata about archival records by reference number, keywords, or topics.

`Integration:` Embed archival data into educational or research platforms.

`Metadata Analysis:` Automate analysis of archival descriptions and linked datasets.

`Digital Access:` Retrieve links to digitized records, where available.

**Technical description**

A RESTful API

The Discovery API allows developers to programmatically query the [Discovery](https://https://discovery.nationalarchives.gov.uk/) service's search engine and database, returning results in XML or JSON formats for further processing. This enables access to over 32 million record descriptions held by The National Archives and more than 2,500 other archives and institutions across the UK.

Please note that the API is offered as a beta service, with some functionality still under development.

```python
# Introduction to Claude 3.5 Sonnet API
# Prerequisites: An Anthropic API key is required

# First, install the Anthropic Python library
!pip install anthropic

# Import required libraries
import anthropic
import os
from google.colab import userdata

# Set up your API key securely
# Note: Store your API key in Colab's secrets manager
api_key = userdata.get('api-practice-key')  # Students should add their key in Colab's secrets
client = anthropic.Anthropic(api_key=api_key)

def ask_claude(question, system_prompt=None):
    """
    Function to send a question to Claude and get a response

    Args:
        question (str): The question to ask Claude
        system_prompt (str, optional): System prompt to guide Claude's behavior

    Returns:
        str: Claude's response
    """
    # Create message parameters
    params = {
        "model": "claude-3-5-sonnet-20241022",
        "max_tokens": 1000,
        "messages": [
            {
                "role": "user",
                "content": question
            }
        ]
    }

    # Add system prompt if provided
    if system_prompt:
        params["system"] = system_prompt

    # Create the message
    message = client.messages.create(**params)

    return message.content[0].text

# Example usage with system prompt
system_prompt = "You are a cultural anthropologist with an interest in mesomerican cultures. Provide detailed, academic-style responses with specific dates and cultural context. You always provide academic references and a bibliography in your answers"
history_question = "What was daily life like in the pueblo cultures of the south-west United States around 1000 AD (pre-contact)?"

# Get response with system prompt
response = ask_claude(history_question, system_prompt)
print("Response with system prompt:")
print(response)

# Get response without system prompt
response_no_system = ask_claude(history_question)
print("\nResponse without system prompt:")
print(response_no_system)

# Exercise for students:
# 1. Try different system prompts to guide Claude's behavior
# 2. Compare responses with and without system prompts
# 3. Experiment with different parameters like max_tokens
# 4. Create specialized system prompts for different subjects
```

```python
# First, install the Anthropic Python library
!pip install anthropic

import anthropic
import os
from google.colab import userdata
from time import sleep
import pandas as pd

def ask_claude(question, model="claude-3-5-sonnet-20241022", temperature=0.7, max_tokens=1000, system=""):
    """
    Enhanced function to interact with Claude with configurable parameters
    """
    try:
        client = anthropic.Anthropic(api_key=userdata.get('api-practice-key'))
        message = client.messages.create(
            model=model,
            max_tokens=max_tokens,
            temperature=temperature,
            system=system,
            messages=[{"role": "user", "content": question}]
        )
        return message.content[0].text
    except Exception as e:
        return f"Error: {str(e)}"

# Experiment 1: Temperature Comparison
def experiment_temperature():
    question = "Write a short story about a robot learning to dance"
    temperatures = [0.0, 0.5, 1.0]

    print("Experiment 1: Temperature Effects\n")
    for temp in temperatures:
        print(f"\nTemperature = {temp}")
        print("-" * 50)
        print(ask_claude(question, temperature=temp))
        sleep(2)  # Avoid rate limiting

# Experiment 2: System Message Impact
def experiment_system_prompts():
    question = "Tell me about dolphins"
    systems = [
        "You are a marine biologist",
        "You are explaining to a 5-year-old",
        "You are a poet"
    ]

    print("\nExperiment 2: System Message Effects\n")
    for system in systems:
        print(f"\nSystem: {system}")
        print("-" * 50)
        print(ask_claude(question, system=system))
        sleep(2)

# Experiment 3: Token Length Impact
def experiment_token_length():
    question = "Explain quantum computing"
    tokens = [100, 300, 500]

    print("\nExperiment 3: Token Length Effects\n")
    for token in tokens:
        print(f"\nMax Tokens: {token}")
        print("-" * 50)
        print(ask_claude(question, max_tokens=token))
        sleep(2)

# Run experiments
if __name__ == "__main__":
    print("Starting Claude API Experiments...")
    experiment_temperature()
    experiment_system_prompts()
    experiment_token_length()
```

```python
# Install required libraries
!pip install anthropic ipywidgets

# Import required libraries
import anthropic
import os
import re
from google.colab import userdata
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets
from datetime import datetime

# Set up your API key securely using userdata
api_key = userdata.get('api-practice-key')
client = anthropic.Anthropic(api_key=api_key)

# Create input widgets for the raw deposition data
case_name_widget = widgets.Text(
    description='Case Name:',
    placeholder='Enter case name',
    layout={'width': 'auto'}
)

proctors_widget = widgets.Text(
    description='Proctors:',
    placeholder='Enter proctors',
    layout={'width': 'auto'}
)

date_widget = widgets.Text(
    description='Date:',
    placeholder='Enter deposition date',
    layout={'width': 'auto'}
)

folio_widget = widgets.Text(
    description='Folio Ref:',
    placeholder='Enter folio reference',
    layout={'width': 'auto'}
)

special_focus_widget = widgets.Text(
    description='Special Focus:',
    placeholder='Enter special focus',
    layout={'width': 'auto'}
)

deposition_widget = widgets.Textarea(
    description='Deposition:',
    placeholder='Enter full deposition text here',
    layout={'width': '95%', 'height': '400px'}
)

# Create submit button and output area
submit_button = widgets.Button(
    description='Analyze Document',
    button_style='success',
    layout={'width': 'auto'}
)

output_area = widgets.Output()

def analyze_document(b):
    """Function to run when submit button is clicked"""
    with output_area:
        clear_output()
        print("Starting analysis...")

        # Construct input text
        input_text = f"""<case_name>{case_name_widget.value}</case_name>
<proctors>{proctors_widget.value}</proctors>
<deposition-date>{date_widget.value}</deposition-date>
<folio_ref>{folio_widget.value}</folio_ref>
<special_focus>{special_focus_widget.value}</special_focus>
<deposition_text>
{deposition_widget.value}
</deposition_text>"""

        try:
            response = analyze_historical_document(input_text)

            # Save to CSV file with timestamp
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            filename = f'historical_analysis_{timestamp}.csv'
            with open(filename, 'w', encoding='utf-8') as f:
                f.write(response)

            print(f"\nAnalysis complete! Results saved to '{filename}'")
            print("\nResponse from Claude:")
            print(response)

        except Exception as e:
            print(f"An error occurred: {str(e)}")

def analyze_historical_document(input_text):
    """Process historical document and get analysis from Claude"""
    print("\nStarting document analysis...")

    system_message = """<system-prompt>
You are an expert historical analyst specializing in 17th-century maritime legal documents. Your task is to analyze a deposition from the High Court of Admiralty and provide a comprehensive analysis in a single CSV (Comma Separated Values) file.
</system-prompt>

<instructions>
Instructions:
1. Carefully read and analyze the entire document.
2. Extract all relevant information according to the framework below.
3. Create an entry in the CSV for EVERY variable listed in the framework.
4. When data is not available for a variable, explicitly state "not specified".
5. When you refer to specific quotes from the text, assign them a number and include the full quoted text in a dedicated section at the end of your analysis.
6. Organize your analysis into a single CSV file with these exact columns:
   Section,Subsection,Category,Description,Quote Number
7. Represent the hierarchical structure of the framework using the Section and Subsection columns.
8. For entries that don't fit neatly into the Category or Description columns, use those columns creatively to convey the information.
9. Maintain consistent detail level across all categories.
10. Flag any uncertainties and ambiguities.
11. Provide confidence levels for all analytical elements in a consolidated table within the CSV.
12. Include numbered quotes and provide the full text of each quote in a QUOTES section at the end of the CSV.
13. All outputs must be provided within one csv code block, with no text outside it.
14. Every single variable from sections 1-10 of the framework must have an entry in the CSV.
</instructions>

<analytical-framework>
1. CORE_DOCUMENT_METADATA
   • Reference number
   • Case name
   • Proctors
   • Date of document
   • Document type
   • Physical condition/completeness
   • Language(s) used

2. REQUIRED_ANALYTICAL_ELEMENTS
   • Key individual details (for each individual):
     - Name, occupation, role in case, age
     - Residence (parish, town, country)
   • Mariner behavior
   • Language skills and literacy
   • Signatures/marks analysis
   2.2 Maritime Assets
       • Key ships
       • Vessel specifications
       • Operational status
   2.3 Commercial Elements
       • Key commodities/materials
       • Financial transactions
       • Contractual undertakings
   2.4 Geographic Framework
       • Key locations
       • Route analysis
       • Jurisdictional boundaries
   2.5 Temporal Elements
       • Chronological sequence
       • Weather events
       • Military events
   2.6 Documentary Analysis
       • Key documents
       • Period-specific language
       • Documentary chain
   2.7 Legal Framework
       • Legal/arbitration questions
       • Legal processes
       • Jurisdictional matters
   2.8 Critical Analysis
       • Discrepancies
       • Biases and limitations
       • Key findings

3. LOCATION_TAGGING_SYSTEM
   3.1 Ports: departure, arrival, intermediate
   3.2 Maritime Spaces: sea areas, coastal zones, river systems
   3.3 Legal Venues: court locations, jurisdiction areas
   3.4 Commercial Places: trading posts, merchant locations

4. EVENT_CLASSIFICATION
   4.1 Maritime Operations: loading, voyage, incident, weather events
   4.2 Legal Proceedings: court, testimony, decision events
   4.3 Commercial Activities: trade, financial, contract events

5. DETAILED_CATEGORIZATION
   5.1 Legal proceedings
   5.2 Vessel information
   5.3 Cargo details
   5.4 Personnel information
   5.5 Commercial transactions
   5.6 Maritime operations
   5.7 Key documents

6. CROSS-REFERENCE_MATRIX
   Relationships between key elements identified in the document

7. CONFIDENCE_LEVELS
   • High confidence elements
   • Medium confidence elements
   • Low confidence elements
   • Speculative elements

8. SUMMARY_SECTIONS
   8.1 Executive Summary
   8.2 Key Findings
   8.3 Areas for Further Research
   8.4 Notable Uncertainties
   8.5 Historical Context

9. NARRATIVE_OVERVIEW
   Brief narrative (200 words) synthesizing:
   - Key events and timeline
   - Main participants and their roles
   - Critical legal/commercial issues
   - Most significant findings
   - Historical significance

10. METADATA
    • Analysis version
    • Date of analysis
    • Analyst details
    • Sources consulted
    • Review status
</analytical-framework>

<csv-output-example>
Example of CSV output:
Section,Subsection,Category,Description,Quote Number
CORE_DOCUMENT_METADATA,Reference number,Folio,HCA 13/70 ff.430v-431r,1
REQUIRED_ANALYTICAL_ELEMENTS,Personnel Analysis,Key individuals [KI],"James Chadwick, Mariner, witness",2
QUOTES,1,Full Text,"[The actual quote from the text]",
QUOTES,2,Full Text,"[The actual quote from the text]",
</csv-output-example>"""

    user_message = f"""Analyze this historical document according to the complete analytical framework. Important:
1. Include an entry for EVERY variable listed in the framework
2. Use "not specified" explicitly where no information is available
3. Include numbered quotes and their full text in a QUOTES section
4. Ensure complete coverage of all sections 1-10

{input_text}

Remember that all variables must be addressed, even if only to note "not specified". No variable from the framework should be omitted from the analysis."""

    print("\nSending request to Claude...")

    try:
        message = client.messages.create(
            model="claude-3-5-sonnet-20241022",
            max_tokens=4000,
            system=system_message,
            messages=[
                {
                    "role": "user",
                    "content": user_message
                }
            ]
        )

        # Process response to ensure QUOTES section is included
        response = message.content[0].text
        if "QUOTES" not in response:
            print("\nWarning: No QUOTES section found in response")
        return response

    except Exception as e:
        print(f"Error during Claude API call: {str(e)}")
        raise

# Connect the button to the function
submit_button.on_click(analyze_document)

# Display the form
display(HTML("<h2>Historical Document Analysis</h2>"))
display(HTML("<p>Enter the deposition details below. Analysis will include quotes from the source text.</p>"))
display(case_name_widget)
display(proctors_widget)
display(date_widget)
display(folio_widget)
display(special_focus_widget)
display(deposition_widget)
display(submit_button)
display(output_area)
```

## **7.4 Processor module**

The analytical engine in the prompt

```python
!pip install anthropic requests beautifulsoup4 ipywidgets tqdm

import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import anthropic
import requests
import re
from bs4 import BeautifulSoup
from typing import List
import json
from datetime import datetime
from google.colab import userdata
from tqdm.notebook import tqdm
import traceback

# Create simple widgets
url_input = widgets.Text(
    value='https://github.com/Addaci/HCA/blob/main/HCA_13_58_Full_Volume_Processed_Text_EDITED_Ver.1.0_25062024.txt',
    description='GitHub URL:',
    layout={'width': '90%'}
)

chunk_size = widgets.IntSlider(
    value=12000,
    min=1000,
    max=15000,
    step=1000,
    description='Chunk Size:'
)

num_chunks = widgets.IntSlider(
    value=1,
    min=1,
    max=50,
    step=1,
    description='Number of Chunks:'
)

analysis_types = widgets.SelectMultiple(
    options=['Key Individuals', 'Ships and Maritime Assets', 'Geographic Locations',
             'Dates and Timeline', 'Legal Proceedings', 'Commercial Activities',
             'Weather Events', 'Military Actions', 'Document Structure'],
    value=['Key Individuals', 'Ships and Maritime Assets', 'Geographic Locations'],
    description='Analysis Types:',
    layout={'width': 'auto', 'height': '200px'}
)

output_format = widgets.Dropdown(
    options=['CSV', 'Text', 'JSON'],
    value='Text',
    description='Output Format:'
)

concatenate = widgets.Checkbox(
    value=True,
    description='Concatenate Results'
)

analyze_button = widgets.Button(
    description='Analyze Document',
    button_style='success'
)

output = widgets.Output()  # Create the output widget

# Display widgets
display(HTML("<h2>HCA Document Analysis</h2>"))
display(url_input)
display(chunk_size)
display(num_chunks)
display(analysis_types)
display(output_format)
display(concatenate)
display(analyze_button)
display(output)  # Make sure to display the output widget


class HCAProcessor:
    def __init__(self, api_key):
        self.client = anthropic.Anthropic(api_key=api_key)

    def fetch_content(self, url):
        # Convert to raw GitHub URL
        raw_url = url.replace('github.com', 'raw.githubusercontent.com').replace('/blob/', '/')
        response = requests.get(raw_url)
        return response.text

    def preprocess_text(self, text):
        # Clean the text
        text = BeautifulSoup(text, 'html.parser').get_text()
        text = re.sub(r'\s+', ' ', text)
        return text.strip()

    def split_into_chunks(self, text, chunk_size):
        chunks = []
        for i in range(0, len(text), chunk_size):
            chunk = text[i:i + chunk_size]
            # Adjust chunk to end at a sentence
            if i + chunk_size < len(text):
                last_period = chunk.rfind('.')
                if last_period != -1:
                    chunk = chunk[:last_period + 1]
            chunks.append(chunk)
        return chunks

    def analyze_chunk(self, chunk, analysis_types):
        prompt = self.create_prompt(chunk, analysis_types)
        response = self.client.messages.create(
            model="claude-3-5-sonnet-20241022",
            max_tokens=4000,
            system="You are an expert in analyzing historical maritime documents.",
            messages=[{"role": "user", "content": prompt}]
        )
        return response.content[0].text

    def create_prompt(self, text, analysis_types):
        prompt = f"""Analyze this historical document excerpt focusing on the following aspects:

{text}

Analysis requested for:
"""
        for analysis_type in analysis_types:
            prompt += f"- {analysis_type}\n"
        return prompt


def process_document(b):
    # Clear the output directly
    output.clear_output()

    try:
        print("Starting analysis...")

        # Get API key
        api_key = userdata.get('api-practice-key')
        if not api_key:
            print("Error: API key not found in Colab secrets")
            return

        # Initialize processor
        processor = HCAProcessor(api_key)

        # Fetch content
        print("Fetching document...")
        content = processor.fetch_content(url_input.value)

        # Preprocess
        print("Preprocessing text...")
        text = processor.preprocess_text(content)

        # Split into chunks
        print("Splitting into chunks...")
        chunks = processor.split_into_chunks(text, chunk_size.value)
        chunks = chunks[:num_chunks.value]  # Only process requested number of chunks

        # Process chunks with progress bar
        results = []
        for i, chunk in enumerate(tqdm(chunks, desc="Analyzing chunks")):
            result = processor.analyze_chunk(chunk, analysis_types.value)
            results.append(result)

        # Save results
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f'hca_analysis_{timestamp}.{output_format.value.lower()}'

        if concatenate.value:
            if output_format.value == 'JSON':
                with open(filename, 'w') as f:
                    json.dump({'results': results}, f, indent=2)
            else:
                with open(filename, 'w') as f:
                    f.write('\n\n---\n\n'.join(results))
        else:
            for i, result in enumerate(results):
                chunk_filename = f'hca_analysis_chunk_{i+1}_{timestamp}.{output_format.value.lower()}'
                with open(chunk_filename, 'w') as f:
                    if output_format.value == 'JSON':
                        json.dump({'result': result}, f, indent=2)
                    else:
                        f.write(result)

        print(f"Analysis complete! Results saved to {filename}")

    except Exception as e:
        print(f"Error: {str(e)}")
        traceback.print_exc()  # Print the full traceback for debugging

# Connect the button to the processing function
analyze_button.on_click(process_document)
```
